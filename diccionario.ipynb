```python
#Librerias
import pandas as pd

# Hacer diccionario con los nombres de las personas primera persona = 1, 2, etc
data = pd.read_csv("/Users/joshchaidez/Globant_Motivation_Prediction/data/data_globant.csv")
personas = {nombre: i+1 for i, nombre in enumerate(data["Name"].unique())}
    
# Guardar dataset modificado etiqueta/nombre
data["PersonID"] = data["Name"].map(personas)
data.to_csv("/Users/joshchaidez/Globant_Motivation_Prediction/data/data_modificado.csv", index=False)

# data modificado head 
print(data.head())

# Drop todo menos nombre y personaID
data_modificado = data[["Name", "PersonID"]]
data_modificado.to_csv("/Users/joshchaidez/Globant_Motivation_Prediction/data/data_nombres_personaID.csv", index=False)

# Eliminar duplicados
data_nombres_personaID = data_modificado.drop_duplicates()
data_nombres_personaID.to_csv("/Users/joshchaidez/Globant_Motivation_Prediction/data/data_nombres_personaID_sin_duplicados.csv", index=False)
```
